### Import library yang akan digunakan

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
df.head(10)
```

## 1.Grafik Batang (Bar Chart)

Karena jumlah dataset saya terlalu banyak saya hanya mengambil data dari range 0-200 pada grafik batang

```python
plt.figure(figsize=(20,15))
Lokasi = df.loc[0:200,"Location"]
Total_Deaths = df.loc[0:200,"Total Deaths"]
x=list(Lokasi)
y=list(Total_Deaths)
plt.title("Total Kematian Pada Suatu Daerah",fontsize=30)
plt.xlabel("Total Kematian", fontsize=30)
plt.ylabel("Lokasi Daerah",fontsize=30)
plt.barh(x,y,color="red")
plt.show()
```

### Kesimpulan yang didapat dari grafik diatas yaitu bahwa dari data yang kita inputkan, daerah yang paling banyak terkena wabah covid adalah DKI Jakarta dengan total kematian lebih dari 40 jiwa, dan Jawa Timur dengan total kematian mendekati 40 jiwa

## 2.Grafik Garis

```python
df["Date"]=pd.to_datetime(df["Date"])
df.dtypes
year_deaths=df.resample('Y',on='Date').mean()
year_deaths.head()
```

```python
plt.figure(figsize=(15,15))
sns.lineplot(data=year_deaths,x=year_deaths.index,y='Total Deaths')
plt.xlabel("Keterangan Date",fontsize=20)
plt.ylabel("Total Kematian",fontsize=20)
plt.title("Total Kematian Pertahun",fontsize=30)
plt.show()
```

### Kesimpulan yang didapat dari Grafik diatas yaitu jumlah kematian dari tahun 2021 sampai tahun 2022 sangat banyak, sampai menyentuh angka 8000 lebih jiwa yang meninggal karena wabah covid 19

## 3.Grafik Pie

Karena data terlalu banyak,disini saya mengambil range tertentu untuk membentuk suatu grafik pie

Untuk data Location,saya hanya menggunakan data dari lokasi Jawa Timur, Kalimantan Timur, Jawa Barat, Jawa Tengah, Bali, dan Banten

```python
lokasi=df[df["Location"].isin(["Jawa Timur","Kalimantan Timur","Jawa Barat","Bali","Banten"])]
data=lokasi["Location"].value_counts().rename_axis("Lokasi").reset_index(name="Id_Count")
data.head()
```

```python
plt.pie(
    data["Id_Count"],
    labels=data["Lokasi"],
    startangle=90,autopct='%.1f%%',
    shadow=True)
plt.title("Banyak Kota Yang Muncul Dalam Dataset")
plt.show()
```

## Kesimpulan yang didapat dari grafik diatas yaitu berapa persen sebuah kota muncul didalam dataset

## Saya juga menambahkan pie chart lain yang dimana pie chart ini digunakan untuk melihat berapa persen total kasus yang terjadi di Indonesia pada tahun 2021

```python
df_pie=df[df["Date"].dt.strftime('%Y')=='2021']
df_pie=df_pie.groupby(df_pie.Date.dt.strftime('%m')).sum().reset_index()
df_pie.head()
```

```python
plt.figure(figsize=(15,15))
plt.pie(df_pie["Total Cases"],labels=df_pie["Date"],startangle=90,autopct='%1.2f%%')
plt.title("Kasus Covid 19 2021 di Indonesia",fontsize=20,color='Blue')
plt.show()
```

## Kesimpulan dari grafik diatas adalah total kasus di Indonesia dari bulan Januari sampai Desember pada tahun 2021 selalu meningkat setiap bulannya walau sempat turun sedikit pada bulan November, tetapi kembali naik lagi pada bulan Desember

# 4.Scatter Plot

## Membandingkan total kematian dan total sembuh

```python
deaths=df.loc[0:1000,"Total Deaths"]
recovered=df.loc[0:1000,"Total Recovered"]
plt.figure(figsize=(17,17))
plt.title("Membandingkan Jumlah Meninggal Dan Jumlah Sembuh Karena Covid 19", fontsize=20)
sns.set_style("darkgrid")
sns.scatterplot(data=df,x=deaths,y=recovered,hue='Location')
plt.xlabel("Total Meninggal Dunia",fontsize=15)
plt.ylabel("Total Sembuh",fontsize=15)
plt.show()
```

## Dari grafik diatas kita dapat membandingkan jumlah meninggal dunia dengan jumlah sembuh dari kasus covid 19 di setiap Lokasi yang ada pada dataset

# 5.Box Plot
untuk melihat penyebaran data dengan atribut kategorikal terhadap data dengan atribut numerik (Category Vs Numerik)

```python
Lokasi = df.loc[0:1000,"Province"]
Total_Cases = df.loc[0:1000,"Total Deaths"]
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x=Lokasi,y=Total_Cases)
plt.xticks(rotation='vertical')
plt.show()
```
