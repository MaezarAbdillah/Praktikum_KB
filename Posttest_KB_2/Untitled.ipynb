#### Informasi yang didapat dari Covid 19 Indonesia Datasets

```python
info ={
    'total_record':0,
    'atribut': [] ,
    'atribut_object':0,
    'atribut_number':0,
    'null_atribut':0
}

info['total_record'] = df.shape[0]
info['atribut'] = list(df.columns.values)
info['atribut_object'] = list(df.select_dtypes(['object']).columns)
info['atribut_number'] = list(df.select_dtypes(['float64','int64']).columns)
info['null_atribut'] = list(df.loc[:, df.isnull().any()].columns)
```

#### Jumlah Record

```python
print("Jumlah Record : ", info['total_record'])
```

#### Nama Atribut dan Jumlah Atribut

```python
print("Nama-nama Atribut : ", info['atribut'])
print("Jumlah Atribut : ", len(info['atribut']))
```

#### Jumlah Atribut dengan Tipe Objek

```python
print("Nama-nama Atribut dengan tipe data object : ", info['atribut_object'])
print("Jumlah Atribut dengan tipe data Object : ", len(info['atribut_object']))
```

#### Jumlah Atribut dengan Tipe Data Angka

```python
print("Nama-nama Atribut dengan tipe data Angka : ", info['atribut_number'])
print("Jumlah Atribut dengan Tipe Data Angka : ", len(info['atribut_number']))
```

#### Jumlah Atribut yang memiliki nilai null

```python
print("Nama-nama Atribut yang memiliki nilai null : ", info['null_atribut'])
print("Jumlah Atribut yang memiliki nilai null : ", len(info['null_atribut']))
```
